**Coorelation Calculation** <br>
Generates the coorelation heatmap/graph for every year. Sums every year and averages, to create a heatmap containing the averaged coorelations from 2003 to 2017

```python
num_features = 18
features = ['pixel_id', 'Road_Dist', 'Elec_Dist', 'Cultivated_Prop', 'aet__maximum_Normal', 'aet__mean_Normal','aet__minimum_Normal', 'cwd__maximum_Normal', 'cwd__mean_Normal','cwd__minimum_Normal', 'pck__maximum_Normal', 'pck__mean_Normal','pck__minimum_Normal', 'aet__mean_ThreeYear_Dev', 'cwd__mean_ThreeYear_Dev', 'Mean_Housing_Dens_25km', 'FFWI', 'value']

temp = [0] * num_features
arr2D = [temp] * num_features

for n in range(num_features):
   arr2D[n] += df_combined_2003.corr()[features[n]].to_numpy()
   arr2D[n] += df_combined_2004.corr()[features[n]].to_numpy()
   arr2D[n] += df_combined_2005.corr()[features[n]].to_numpy()
   arr2D[n] += df_combined_2006.corr()[features[n]].to_numpy()
   arr2D[n] += df_combined_2007.corr()[features[n]].to_numpy()
   arr2D[n] += df_combined_2008.corr()[features[n]].to_numpy()
   arr2D[n] += df_combined_2009.corr()[features[n]].to_numpy()
   arr2D[n] += df_combined_2010.corr()[features[n]].to_numpy()
   arr2D[n] += df_combined_2011.corr()[features[n]].to_numpy()
   arr2D[n] += df_combined_2012.corr()[features[n]].to_numpy()
   arr2D[n] += df_combined_2013.corr()[features[n]].to_numpy()
   arr2D[n] += df_combined_2014.corr()[features[n]].to_numpy()
   arr2D[n] += df_combined_2015.corr()[features[n]].to_numpy()
   arr2D[n] += df_combined_2016.corr()[features[n]].to_numpy()
   arr2D[n] += df_combined_2017.corr()[features[n]].to_numpy()
   arr2D[n] = arr2D[n]/15
```

```python
plt.figure(figsize=(12,8))
plt.title("Feature-Feature Coorelation")
sns.heatmap(arr2D,annot=True,fmt='.2f',xticklabels=features,yticklabels=features)
plt.show()
```
